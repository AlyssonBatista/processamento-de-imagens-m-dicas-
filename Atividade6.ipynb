## Imagens do método laplaciano e depois binarização pelo método de otsu

```python
from skimage.filters import (threshold_otsu, threshold_triangle, threshold_niblack, threshold_sauvola)
from skimage import img_as_ubyte

 # Declare the variables we are going to use
ddepth = cv2.CV_16S
kernel_size = 7
window_name = "Laplace Demo"

img = cv2.imread('imagem3.jpg')

# Remove noise by blurring with a Gaussian filter
src = cv2.GaussianBlur(img, (3, 3), 0)

gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# laplac = cv2.Canny(gray,100,200)
laplac = cv2.Laplacian(gray, ddepth, ksize=kernel_size)
# laplac = cv2.Laplacian(gray,cv2.CV_64F)


thresh_otsu = threshold_otsu(laplac)
binary_otsu = laplac > thresh_otsu # sempre que o fundo do objeto ficar branco inverta o sinal que está entre laplac e thresh_otsu 
binary_otsu = img_as_ubyte(binary_otsu)

# Apply Laplace function
#laplaciano
# laplacian = np.uint8(laplacian)


# converting back to uint8
laplaciano = cv2.convertScaleAbs(binary_otsu)

imagem_BGR = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

cv2_imshow(laplaciano)
```

## 1 - Dilatação

```python
# Rectangular Kernel
# kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(9,9))
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

# Elliptical Kernel
# kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(9,9))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

# Cross-shaped Kernel
# kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(9,9))
kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

dilation1 = cv2.dilate(laplaciano,kernel1,iterations = 1)
dilation2 = cv2.dilate(laplaciano,kernel2,iterations = 1)
dilation3 = cv2.dilate(laplaciano,kernel3,iterations = 1)

fig,ax = plt.subplots(1,4,figsize=(25,15))

ax[0].set_title('Laplaciano',fontsize=19)
ax[0].imshow(laplaciano,'gray')
ax[1].set_title('Dilatação com el. estrut. Retangulo',fontsize=19)
ax[1].imshow(dilation1,'gray')
ax[2].set_title('Dilatação com el. estrut. elipse',fontsize=19)
ax[2].imshow(dilation2,'gray')
ax[3].set_title('Dilatação com el. estrut. Cruz',fontsize=19)
ax[3].imshow(dilation3,'gray')
plt.show()
```

## 2 - Erosão

```python
# Rectangular Kernel
# kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

# Elliptical Kernel
# kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

# Cross-shaped Kernel
# kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

erosao1 = cv2.erode(laplaciano,kernel1,iterations = 1)
erosao2 = cv2.erode(laplaciano,kernel2,iterations = 1)
erosao3 = cv2.erode(laplaciano,kernel3,iterations = 1)

fig,ax = plt.subplots(1,4,figsize=(25,15))

ax[0].set_title('Laplaciano',fontsize=19)
ax[0].imshow(laplaciano,'gray')
ax[1].set_title('Erosão com el. estrut. Retangulo',fontsize=19)
ax[1].imshow(erosao1,'gray')
ax[2].set_title('Erosão com el. estrut. elipse',fontsize=19)
ax[2].imshow(erosao2,'gray')
ax[3].set_title('Erosão com el. estrut. Cruz',fontsize=19)
ax[3].imshow(erosao3,'gray')
plt.show()

```

## 4 - Detecção de bordas

```python
#Rectangular Kernel
# kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

# Elliptical Kernel
# kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

# Cross-shaped Kernel
# kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

bordas1 = cv2.morphologyEx(laplaciano, cv2.MORPH_GRADIENT, kernel1)
bordas2 = cv2.morphologyEx(laplaciano, cv2.MORPH_GRADIENT, kernel2)
bordas3 = cv2.morphologyEx(laplaciano, cv2.MORPH_GRADIENT, kernel3)

 

fig,ax = plt.subplots(1,4,figsize=(25,15))

ax[0].set_title('Laplaciano',fontsize=19)
ax[0].imshow(laplaciano,'gray')
ax[1].set_title('Bordas com el. estrut. Retangulo',fontsize=19)
ax[1].imshow(bordas1,'gray')
ax[2].set_title('Bordas com el. estrut. elipse',fontsize=19)
ax[2].imshow(bordas2,'gray')
ax[3].set_title('Bordas com el. estrut. Cruz',fontsize=19)
ax[3].imshow(bordas3,'gray')
plt.show() 
```

## 5 - Abertura e Fechamento

### obs: Abeertura é um nome dado ao processo de erosão seguido de dilatação (utilizando o mesmo elemento estruturante). É útil na remoção de ruído da imagem.

```python

#Rectangular Kernel
# kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

# Elliptical Kernel
# kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

# Cross-shaped Kernel
# kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

abertura1 = cv2.morphologyEx(laplaciano, cv2.MORPH_OPEN, kernel1)
abertura2 = cv2.morphologyEx(laplaciano, cv2.MORPH_OPEN, kernel2)
abertura3 = cv2.morphologyEx(laplaciano, cv2.MORPH_OPEN, kernel3)

fig,ax = plt.subplots(1,4,figsize=(25,15))

ax[0].set_title('Laplaciano',fontsize=19)
ax[0].imshow(laplaciano,'gray')
ax[1].set_title('Abertura com el. estrut. Retangulo',fontsize=19)
ax[1].imshow(abertura1,'gray')
ax[2].set_title('Abertura com el. estrut. elipse',fontsize=19)
ax[2].imshow(abertura2,'gray')
ax[3].set_title('Abertura com el. estrut. Cruz',fontsize=19)
ax[3].imshow(abertura3,'gray')
plt.show() 
```

### O Closing é o contrário do opening, ou seja, é uma dilatação seguida por erosão. Ele pode ser utilizado para fechar pequenos buracos dentro dos objetos em primeiro plano ou pequenos pontos pretos na imagem.

```python
#Rectangular Kernel
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
# kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

# Elliptical Kernel
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
# kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

# Cross-shaped Kernel
kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
# kernel3 = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))

fechamento1 = cv2.morphologyEx(laplaciano, cv2.MORPH_CLOSE, kernel1)
fechamento2 = cv2.morphologyEx(laplaciano, cv2.MORPH_CLOSE, kernel2)
fechamento3 = cv2.morphologyEx(laplaciano, cv2.MORPH_CLOSE, kernel3)

fig,ax = plt.subplots(1,4,figsize=(25,15))

ax[0].set_title('Laplaciano',fontsize=19)
ax[0].imshow(laplaciano,'gray')
ax[1].set_title('Fechamento com el. estrut. Retangulo',fontsize=19)
ax[1].imshow(fechamento1,'gray')
ax[2].set_title('Fechamento com el. estrut. elipse',fontsize=19)
ax[2].imshow(fechamento2,'gray')
ax[3].set_title('Fechamento com el. estrut. Cruz',fontsize=19)
ax[3].imshow(fechamento3,'gray')
plt.show() 
 
```
