## 1 - Estudar a equação que passa de RGB para tons de cinza -- testar e comparar usando a luminosidade e a média 

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('imagem9.jpg')
print(img.shape)
#precisamos transformar isso para que o Matplotlib leia corretamente
imagem_BGR = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(imagem_BGR)

# extrair os três canais
R, G, B = imagem_BGR[:,:,0], imagem_BGR[:,:,1],imagem_BGR[:,:,2]
 
```

### Método médio 
Tons de cinza = (R + G + B / 3)

```python
medio =  (R + G + B)/3
```

### método de luminosidade

Nova imagem em tons de cinza = 0,3 * R + 0,59 * G + 0,11 * B

```python
luminosidade = 0.3 * R + 0.59 * G + 0.11 * B
```

```python
cv2_imshow( np.hstack((medio,luminosidade)))
```

## 2 - Converter as imagens de RGB para HSI

```python
imagemhsv = cv2.cvtColor(imagem_BGR, cv2.COLOR_BGR2HSV)
cv2_imshow(imagemhsv)
```

## 3 - Identificar as cores que tem na imagem (escolhendo as doze cores principais e apresentar de forma de histograma)

```python
img = cv2.imread("imagem9.jpg")
# Converte para HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# bola vermelha 
upper_red= np.array([179,255,255])
lower_red  = np.array([174,0,0])

mask1= cv2.inRange(hsv, lower_red, upper_red)
masc_red = cv2.bitwise_and(img,img,mask=mask1)
red = cv2.cvtColor(masc_red, cv2.COLOR_BGR2RGB)

# bola amarela 
upper_amarelo= np.array([27,255,255])
lower_amarelo= np.array([20,50,0])

mask2= cv2.inRange(hsv, lower_amarelo, upper_amarelo)
masc_red = cv2.bitwise_and(img,img,mask=mask2)
amarelo = cv2.cvtColor(masc_red, cv2.COLOR_BGR2RGB)

# bola cinza
upper_cinza= np.array([0,0,200])
lower_cinza= np.array([0,0,0])

mask3= cv2.inRange(hsv, lower_cinza, upper_cinza)
masc_cinza = cv2.bitwise_and(img,img,mask=mask3)
cinza = cv2.cvtColor(masc_cinza, cv2.COLOR_BGR2RGB)

# bola roxo
upper_roxo= np.array([170,200,230])
lower_roxo= np.array([125,0,70])

mask4= cv2.inRange(hsv, lower_roxo, upper_roxo)
masc_roxo = cv2.bitwise_and(img,img,mask=mask4)
roxo = cv2.cvtColor(masc_roxo, cv2.COLOR_BGR2RGB)

# bola rosa
upper_rosa= np.array([170,230,255])
lower_rosa= np.array([160,100,70])

mask5= cv2.inRange(hsv, lower_rosa, upper_rosa)
masc_rosa = cv2.bitwise_and(img,img,mask=mask5)
rosa = cv2.cvtColor(masc_rosa, cv2.COLOR_BGR2RGB)

# bola laranja
upper_laranja= np.array([17,230,255])
lower_laranja= np.array([7,100,70])

mask6= cv2.inRange(hsv, lower_laranja, upper_laranja)
masc_laranja = cv2.bitwise_and(img,img,mask=mask6)
laranja = cv2.cvtColor(masc_laranja, cv2.COLOR_BGR2RGB)

# bola azul_cla
upper_azul_cla= np.array([107,230,255])
lower_azul_cla= np.array([90,100,70])

mask6= cv2.inRange(hsv, lower_azul_cla, upper_azul_cla)
masc_azul_cla = cv2.bitwise_and(img,img,mask=mask6)
azul_cla = cv2.cvtColor(masc_azul_cla, cv2.COLOR_BGR2RGB)

# bola azul_esc
upper_azul_esc= np.array([112,230,255])
lower_azul_esc= np.array([105,140,70])

mask6= cv2.inRange(hsv, lower_azul_esc, upper_azul_esc)
masc_azul_esc = cv2.bitwise_and(img,img,mask=mask6)
azul_esc = cv2.cvtColor(masc_azul_esc, cv2.COLOR_BGR2RGB)

# bola verde
upper_verde= np.array([40,250,255])
lower_verde= np.array([28,90,100])

mask6= cv2.inRange(hsv, lower_verde, upper_verde)
masc_verde = cv2.bitwise_and(img,img,mask=mask6)
verde = cv2.cvtColor(masc_verde, cv2.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(2, 6, figsize=(30,10)) 

 
ax[0,0].set_title('Imagem Original', fontsize=16)
ax[0,0].hist(hsv[:,:,0].ravel(),179,[1,254]);

ax[0,1].set_title('vermelho', fontsize=16)
ax[0,1].hist(cv2.cvtColor(red, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[0,2].set_title('amarelo', fontsize=16)
ax[0,2].hist(cv2.cvtColor(amarelo, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[0,3].set_title('cinza', fontsize=16)
ax[0,3].hist(cv2.cvtColor(cinza, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[0,4].set_title('roxo', fontsize=16)
ax[0,4].hist(cv2.cvtColor(roxo, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[0,5].set_title('rosa', fontsize=16)
ax[0,5].hist(cv2.cvtColor(rosa, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);

ax[1,0].set_title('laranja', fontsize=16)
ax[1,0].hist(cv2.cvtColor(laranja, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[1,1].set_title('azul claro', fontsize=16)
ax[1,1].hist(cv2.cvtColor(azul_cla, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[1,2].set_title('azul escuro', fontsize=16)
ax[1,2].hist(cv2.cvtColor(azul_esc, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);
ax[1,3].set_title('verde', fontsize=16)
ax[1,3].hist(cv2.cvtColor(verde, cv2.COLOR_BGR2HSV)[:,:,2].ravel(),255,[1,254]);

ax[1,4].set_title('Imagem original', fontsize=16)
ax[1,4].imshow(imagem_BGR)
plt.axis('off')
 
plt.show()
```
